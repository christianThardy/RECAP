```python
import json
from collections import defaultdict

import pandas as pd


# with open("/content/synthetic_prompts_merged_full_structure.json", "r", encoding="utf-8") as f:
with open("/content/synthetic_prompts_merged_full_structure_deduped.json", "r", encoding="utf-8") as f:
    full_data = json.load(f)

# Extract all rows from each category into a single list
all_rows = []
for category, content in full_data.items():
    for row in content.get("_rows", []):
        all_rows.append(row)

df = pd.DataFrame(all_rows)
```

```python
# 1) Identify the seed rows you want to dedupe
seed_mask = df["model_used"].isin(["original_seed", "original_seed_forking"])

# 2) Split the frame into seed and non-seed
seed_df     = df[seed_mask].copy()
non_seed_df = df[~seed_mask]

# 3) Drop duplicates inside the seed block
#     If your prompt text alone is sufficient to identify a duplicate, keep just ["prompt"]
#     Otherwise include other columns (e.g. "category") in the subset list.
seed_df_deduped = seed_df.drop_duplicates(subset=["category", "prompt"], keep="first")

# 4) Re-assemble the cleaned dataframe
df_clean = pd.concat([seed_df_deduped, non_seed_df], ignore_index=True)

# Optional: quick sanity check
print(f"Original rows: {len(df)}")
print(f"Rows after deduping seeds: {len(df_clean)}")
```

```python
# df_clean already exists
nested = defaultdict(lambda: {
    "_rows": []
})

for row in df_clean.to_dict(orient="records"):
    cat = row["category"]
    nested[cat]["_rows"].append(row)

# dump it
with open("synthetic_prompts_merged_full_structure_deduped.json", "w", encoding="utf-8") as f:
    json.dump(nested, f, indent=2, ensure_ascii=False)
```
